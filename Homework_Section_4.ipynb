**Step 3: Data Analysis**

The first question is about free throw attempts per game. This metric can shed some light on the player's court position (Shoorting Guard, Center, Point Forward) and style of game (long-range shooter, slasher), indicating if he is fouled more frequently during the match. 

```python
# Calculating each player free throw attempts per game

FT_per_Game = FreeThrowAttempts / Games

FT_per_Game
```

The code contains a warning because of a NAN value in the dataset. The reason behind it is that Derrick Rose didn't play any games in the 2012-13 season due to a knee injury (torn ACL).

The result is also poorly readable. Rounding the numbers should make the result neat and clear.

```python
#Rounding the results

np.round(FT_per_Game, 2)
```

Now it is time to assess the players free throw accuracy.

```python
# Calculating free throw accuracy

FT_Accuracy = FreeThrows / FreeThrowAttempts

np.round(FT_Accuracy,3) * 100
```

The third aspect is points per attempt of each player **excluding free throws**.

```python
# Points excluding Free Throws

Points_ex_FT = (Points - FreeThrows) / FieldGoals

np.round(Points_ex_FT, 2)
```

**Step 4: Data Visualization**

Calculating matrices as done previously is not enough to gather insights about the data. Plotting the results usually provides those "a-ha" moments that we seek in data analysis.

First, a function for the plot will be created.

```python
# Creating the plot function
def myplot(data, playerlist = Players):
    Col = {'KobeBryant':'Black', 'JoeJohnson':'Red','LeBronJames':'Green','CarmeloAnthony':'Blue','DwightHoward':'Magenta','ChrisBosh':'Yellow','ChrisPaul':'Orange','KevinDurant':'Grey','DerrickRose':'Pink','DwayneWade':'Cyan'}
    Mrk = {'KobeBryant':'.', 'JoeJohnson':',','LeBronJames':'o','CarmeloAnthony':'v','DwightHoward':'^','ChrisBosh':'1','ChrisPaul':'s','KevinDurant':'p','DerrickRose':'h','DwayneWade':'d'}
    
    for name in playerlist:
        plt.plot(data[Pdict[name]], c = Col[name], ls = '--', marker = Mrk[name], ms = '7', label = name)
    plt.legend(loc = 'upper left', bbox_to_anchor = (1,1))
    plt.xticks(list(range(0,10)), Seasons, rotation = 'vertical')
    plt.show()
```

i) Free Throw Attempts per Game

```python
myplot(FT_per_Game)
```

Analysis of (i): 
- Most players demonstrate a downward trend in Free Throw Attempts per Game. A reasonable hypothesis for this situation is that as an athlete gets older, he tends to gradually lose athleticism and aggressiveness. The positive note is that a more experienced player has a higher accuracy in long-range shots due to repetition and more acumen, avoiding too much contact and injury-prone situations. Either way, the player is less fouled and therefore has less free throw attempts.

ii) Free Throw Accuracy

```python
myplot(FT_Accuracy)
```

Analysis of (ii):

- Dwight Howard could be considered an outlier in this sample due to his poor performance in free throws. Most players have a track record of free throw accuracy above 70%. Chris Paul was consistent througout the years and had the best free throw accuracy in 2014. Interestingly, he had few attempts per game compared to other players.

iii) Field Goals Accuracy (excluding Free Throws)

```python
myplot(Points_ex_FT)
```

Analysis of (iii)

- The chart above gives a hint on each player's style. Joe Johnson, for example, seems to be a long-range shooter, which is typical of a Shooting Guard/Small Forward. A different profile of player is represented by Dwight Horward, which plays as a Center Forward, therefore, closer to the basket. It justifies his 2-point average of field goals. Chris Paul, again, showed high levels of consistency and marginal improvement.
